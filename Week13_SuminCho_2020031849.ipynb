Import libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns 
```

```python
df = pd.read_csv('mall_customers.csv')
df
```

Read 'mall_customers.csv' file

```python
df.rename(index=str, columns={'Annual Income (k$)':'Income', 'Spending Score (1-100)':'Score'}, inplace=True)
df
```

```python
X=df.drop(['CustomerID', 'Gender'], axis=1)
X
```

k-Means clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, # 1~11 clusters
                    init='k-means++', 
                    max_iter=300, # moving the centroid 300 times per run
                    random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
plt.title("The elbow method")
plt.axvline(x=5, color='red', label='Optimal k')
plt.axhline(y=50000, color='green')
plt.xlabel('number of clusters')
plt.ylabel('intertia')
plt.legend()
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters=5, random_state=0)
y_kmeans = kmeansmodel.fit_predict(X)
print(y_kmeans)
```

```python
km3 = KMeans(n_clusters=3).fit_predict(X)
print(km3)
```

```python
X['Label']=km3
plt.figure(figsize=(8,6))
sns.scatterplot(x=X['Income'], y=X['Score'], hue=X['Label'],
                palette=sns.color_palette('husl'))
plt.title('KMeans with 3 clusters')
plt.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris.feature_names)
```

```python
iris
```

```python
df = pd.DataFrame({'sepal length (cm)':iris.data[:,0],
                   'petal length (cm)':iris.data[:,2]
                  })
df.head()
```

```python
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, # 1~11 clusters
                    init='k-means++', 
                    max_iter=300, # moving the centroid 300 times per run
                    random_state=2)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)


plt.plot(range(1,11), wcss)
plt.title("The elbow method")
plt.axvline(x=5, color='red', label='Optimal k')
# plt.axhline(y=50000, color='green')
plt.xlabel('number of clusters')
plt.ylabel('intertia')
plt.legend()
plt.show()
```

```python
km3_iris = KMeans(n_clusters=3).fit(df)
print(km3_iris)
df['Label']=km3_iris.labels_
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['sepal length (cm)'], y=df['petal length (cm)'], hue=df['Label'],
                palette=sns.color_palette('husl', 3))
plt.title('KMeans with 3 clusters: IRIS')
plt.show()
```

```python
fig = plt.figure(figsize=(16,6))
ax = fig.add_subplot(121)

sns.swarmplot(x='Label', y='sepal length (cm)', data=df, ax=ax, palette=sns.color_palette('rocket'))
ax.set_title('Cluseters according to sepal length')

ax = fig.add_subplot(122)

sns.swarmplot(x='Label', y='petal length (cm)', data=df, ax=ax, palette=sns.color_palette('rocket'))
ax.set_title('Cluseters according to petal length')
```
